# **Template-Based Feedback Generation**

**Template 1**

```python
# TEMPLATE BASED TELECOM FEEDBACK GENERATOR
# Output: template_feedback_dataset.csv

import random
import pandas as pd

# ZENDS Telecom Services
services = [
    "Mobile Network",
    "Broadband Service",
    "Billing & Payments",
    "Customer Support",
    "Mobile App Issues",
    "Service Activation"
]

# Sentiment distribution
sentiment_counts = {
    "positive":3000,
    "negative":3000,
    "neutral":3000,
    "mixed":1000
}

# Templates
positive_templates = [
    "I am very satisfied with the {service} from ZENDS.",
    "The {service} works perfectly.",
    "Excellent experience with the {service}.",
    "ZENDS provides reliable {service}.",
    "Really happy with the quality of the {service}."
]

negative_templates = [
    "I am facing serious issues with the {service}.",
    "The {service} is extremely unreliable.",
    "Very disappointed with the {service}.",
    "ZENDS needs to improve the {service}.",
    "The {service} keeps failing frequently."
]

neutral_templates = [
    "The {service} is average.",
    "My experience with the {service} is okay.",
    "The {service} works but nothing special.",
    "The {service} is acceptable.",
    "The {service} performance is moderate."
]

mixed_templates = [
    "The {service} works well sometimes but fails occasionally.",
    "Customer support was helpful but the {service} is slow.",
    "The {service} quality is good but inconsistent.",
    "Billing was smooth but the {service} connection is unstable."
]

dataset = []

for sentiment,count in sentiment_counts.items():

    for i in range(count):

        service = random.choice(services)

        if sentiment == "positive":
            text = random.choice(positive_templates)

        elif sentiment == "negative":
            text = random.choice(negative_templates)

        elif sentiment == "neutral":
            text = random.choice(neutral_templates)

        else:
            text = random.choice(mixed_templates)

        feedback = text.format(service=service)

        dataset.append([feedback,sentiment,service])

df = pd.DataFrame(dataset,columns=["feedback","sentiment","service"])

df.to_csv("template_feedback_dataset.csv",index=False)

print("Dataset Generated:",len(df))
df.head()
```

**Template 2**

```python
# SENTIMENT CONTROLLED TELECOM DATASET GENERATOR
# Output: sentiment_controlled_dataset.csv

import random
import pandas as pd

services = [
    "Mobile Network",
    "Broadband Service",
    "Billing & Payments",
    "Customer Support",
    "Mobile App Issues",
    "Service Activation"
]

sentiment_distribution = {
    "positive":3000,
    "negative":3000,
    "neutral":3000,
    "mixed":1000
}

phrases = {

"positive":[
"great experience with",
"very satisfied with",
"excellent performance of",
"really impressed by",
"very reliable"
],

"negative":[
"very disappointed with",
"very disointed with",
"constant problems with",
"extremely poor quality of",
"frustrated with",
"terrible performance of"

],

"neutral":[
"average experience with",
"acceptable performance of",
"the service is okay",
"nothing special about",
"moderate performance of"
],

"mixed":[
"good but sometimes problematic",
"works well but occasionally slow",
"support is good but service unstable",
"inconsistent performance of"
]

}

rows = []

for sentiment,count in sentiment_distribution.items():

    for i in range(count):

        service = random.choice(services)
        phrase = random.choice(phrases[sentiment])

        feedback = f"I had {phrase} the {service} provided by ZENDS Communications."

        rows.append([feedback,sentiment,service])

df = pd.DataFrame(rows,columns=["feedback","sentiment","service"])

df.to_csv("sentiment_controlled_dataset.csv",index=False)

print("Dataset Created:",len(df))
df.head()
```
